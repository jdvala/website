```python
import pyLDAvis.gensim
import pickle
from gensim import corpora
from gensim.models.ldamodel import LdaModel as Lda
pyLDAvis.enable_notebook()
```

```python
with open('/home/jay/ANN_Models/LDA/LDA_Corpus.pickle','rb') as corp:
    corpus=pickle.load(corp)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in corpus]
```

```python
dictionary = corpora.Dictionary.load('/home/jay/ANN_Models/LDA/LDA_Dictonary')
ldamodel = Lda.load('/home/jay/ANN_Models/LDA/LDA_Model')    
```

```python
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```

```python
pyLDAvis.display(vis)
```

```python
# Print topics
# I also know that none of my topic is more than 5 words i will set the num_words argument to 5
for i,topic in enumerate(ldamodel.print_topics(num_topics=32, num_words=5)):
    words = topic[1].split("+")
    print (i, words,"\n")
```
